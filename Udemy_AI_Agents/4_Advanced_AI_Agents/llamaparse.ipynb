```python
document_with_instruction=LlamaParse(
    result_type="markdown",
    parsing_instruction="This is the Apple Annual Report. Make a Summary")\
        .load_data(file_path="form10-k.pdf")
```

```python
fileName="apple_10k_instructions.md"
with open(file=fileName,mode="w") as file:
    file.write(document_with_instruction[0].text)
print(f"File {fileName} created successfully.")
```
